```python
from google.colab import userdata
apikey=userdata.get('YOUTUBE_API')
```

```python
from googleapiclient.discovery import build
youtube=build('youtube','v3',developerKey=apikey)
```

```python
def get_comments(video_id,max_comments=100):
  comments=[]

  request=youtube.commentThreads().list(
      part='snippet',
      videoId=video_id,
      textFormat='plainText',
      maxResults=max_comments
  )

  response=request.execute()

  for res in response['items']:
    comment=res["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
    comments.append(comment)

  return comments

```

```python
video_id="sbD6PK5KRdI"
comments=get_comments(video_id)
```

```python
for i, comment in enumerate(comments[:], start=1):
    print(f"{i}. {comment}")
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
import emoji
```

```python
def data_preprocessing(comment):
  comment=emoji.demojize(comment,delimiters=("",""))
  comment=re.sub(r'_',' ',comment)

  comment=re.sub(r'http\S+|www\S+|https\S+','',comment)
  comment=re.sub(r'[^a-zA-Z]',' ',comment)
  comment=comment.lower()
  '''
  words=nltk.word_tokenize(comment)
  stop_words=set(stopwords.words('english'))
  words=[word for word in words if word not in stop_words]
  '''
  lemmatizer=WordNetLemmatizer()
  words=[lemmatizer.lemmatize(word) for word in words]

  return ' '.join(words)
```

```python
processed_comments = [data_preprocessing(comment) for comment in comments]
```

```python
for i, comment in enumerate(processed_comments[:], start=1):
    print(f"{i}. {comment}")
```

```python
def bert_preprocessing(comment):
  comment=re.sub(r'http\S+|www\S+|https\S+', '', comment)
  comment=emoji.demojize(comment,delimiters=(" "," "))
  comment=re.sub(r'_',' ',comment)
  comment=comment.strip()
  return comment
```

```python
bert_comments=[bert_preprocessing(comment) for comment in comments]
```

```python
for i, comment in enumerate(bert_comments[:], start=1):
    print(f"{i}. {comment}")
```

```python
from transformers import AutoTokenizer,AutoModel
import torch
import numpy as np

model_name='google-bert/bert-base-uncased'
tokenizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModel.from_pretrained(model_name)

def bert_embedding(comment):
  inputs=tokenizer(comment,return_tensors='pt',truncation=True,padding=True)
  with torch.no_grad():
    outputs=model(**inputs)
  return outputs.last_hidden_state.mean(dim=1).squeeze(dim=1).numpy()
```

```python
bert_embedd=numpy.array([bert_embedding(comment) for comment in bert_comments])
bert_embedd=bert_embedd.squeeze()
```
